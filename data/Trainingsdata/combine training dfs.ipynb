```python
from tsai.all import *
from datetime import timedelta
```

```python
files = glob.glob('model input CSVs/*.csv')
files
```

```python
dfs = list()
unique_id = 0
format = '%Y-%m-%d %H:%M:%S'
time_min = timedelta(days=25)
for file in files:
    df = pd.read_csv(file, delimiter=';')
    rows = df.shape[0]
    drop_id = np.where(df.isna())[0]
    print(drop_id)

    cut_id = np.where(np.diff(drop_id)!=1)
    for i in range(len(cut_id)-1):
        time_start = datetime.strptime(df['Time'][cut_id[i]], format)
        time_end = datetime.strptime(df['Time'][cut_id[i+1]], format)
        if time_end-time_start < time_min:
            print(range(cut_id[i], cut_id[i+1]))
    #df['unique_id']
    df = df.drop(drop_id)
    dfs.append(df)
```
